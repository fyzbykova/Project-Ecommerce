### Сформулируем возможную задачу от менеджеров для дальнейшего решения
#### Мы хотим найти суммарное количество всех денег по каждой категории, проранжировать эти категории по убыванию заработанных денег, также посмотреть разбивку по брендам

```python
# посчитаем сумму денег в разбивке по каждой категории
# группируем 
# агрегируем и суммируем
# сортируем значения по убыванию

df \
    .groupby('ytem_category', as_index=False) \
    .aggregate({'selling_price': 'sum'}) \
    .sort_values('selling_price', ascending=False)
```

```python
# и финальное добавим разбивку по брендам

money_by_brands = df \
    .groupby(['ytem_category', 'product_brand'], as_index=False) \
    .aggregate({'selling_price': 'sum'}) \
    .sort_values('selling_price', ascending=False)
```

```python
money_by_brands
```

```python
# результат сохраняем в csv файл

money_by_brands.to_csv('money_by_brands.csv', index=False)
```

```python
# проверим, что при группировке не потерялись данные

money_by_brands.selling_price.sum()
```
